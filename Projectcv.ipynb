```python
import PyPDF2

def extract_text_from_pdf(file_path):
    text = ""
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

# Replace with your uploaded file name
resume_text = extract_text_from_pdf("samplecv.pdf")
print(resume_text)
```

```python
job_description = """
We are looking for a customer service assistant in a retail sports store.
The candidate should have strong communication skills, cash handling experience,
teamwork ability, and good customer interaction skills.
Experience in sports environment is a plus.
"""
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words]
    words = [word for word in words if word not in stop_words]

    return words   # return list instead of string
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
import re
def calculate_match_score(resume_text, job_description):

    resume_words = set(clean_text(resume_text))
    job_words = set(clean_text(job_description))

    matched_words = resume_words.intersection(job_words)
    missing_words = job_words - matched_words

    score = len(matched_words)
    total_job_keywords = len(job_words)
    match_percentage = (score / total_job_keywords) * 100

    return score, total_job_keywords, matched_words, missing_words, match_percentage
```

```python
score, total_job_keywords, matched_words, missing_words, match_percentage = calculate_match_score(resume_text, job_description)

print("Match Percentage:", round(match_percentage, 2), "%")
print("Matched Keywords:", matched_words)
print("Missing Keywords:", missing_words)
```

```python
print("\n--- Resume Improvement Suggestions ---")

for word in missing_words:
    if word not in ["plus", "candidate"]:  # ignore non-skill words
        print(f"- Consider adding or highlighting experience related to '{word}'")
```
